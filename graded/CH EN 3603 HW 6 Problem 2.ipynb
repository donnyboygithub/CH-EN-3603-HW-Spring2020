1. Using Figure 2.4 in SHR, generate a plot of the bubble point pressure (in psia) as a function of temperature (in F) for n-pentane.
2. According to the figure, which is more volatile, isopentane or normal pentane? Does the same hold for n-butane/i-butane and n-butene/i-butene?

# Solution

## part 1

At the bubble point,

$x_i=y_i \Rightarrow K_i=1$

Figure 2.4 in SHR was used to obtain the data points of pressure and temperature for n-pentane.

```python
import numpy as np
import matplotlib.pyplot as plt

P=np.array([1500,500,110,30,26,22,19,16,11]) #psia
T=np.array([600,400,240,140,130,120,110,100,80]) #F

plt.plot(T,P,'r-')
plt.grid()
plt.xlabel('Temperature (degrees F)')
plt.ylabel('Bubble Point Pressure (psia)')
plt.show()
```

## part 2

Drawing a line in similar fassion as in part 1 - from isopentane, through K=1 - it can be seen that isopentane, at the same pressure as n-pentane, will boil at a lower temperature, meaning that

$\boxed{isopentane \space \space is \space \space more \space \space volatile \space \space than \space \space n-pentane.}$

Using the same method

$\boxed{i-butane \space \space is \space \space more \space \space volatile \space \space than n-butane.}$

Using the same method

$\boxed{i-butene \space \space is \space \space more \space \space volatile \space \space than n-butene.}$
