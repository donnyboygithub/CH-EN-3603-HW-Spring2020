# Problem 4

Forty-five kg of a solution of 30 wt% ethylene glycol in water is to be extracted with furfural. Using Figure 4.13, calculate the

(a) minimum kg of solvent

(b) maximum kg of solvent

(c) kg of solvent-free extract and raffinate for 45 kg solvent, and the percentage glycol extracted

(d) maximum purity of glycol in the extract and the maximum purity of water in the raffinate for one stage.

# Solution

## Part a

In the diagram above, $\frac{\overline{MF}}{\overline{SM}}=$ ratio of solvent to feed $\Rightarrow$

```python
MF=0.89 #in
SM=8.65 #in
ratio=MF/SM
print('MF/SM=',ratio)
```

Since we know the mass of the feed=SM=45 kg,

```python
SM=45 #kg
MF=ratio*SM #kg
print(MF,'kg is the minimum mass of solvent.')
```

## Part b

```python
SM=45 #kg
MF=ratio*SM #kg
print(MF,'kg is the maximum mass of solvent.')
```

## Part c

```python
m_solvent=45 #kg
m_solution=45 #kg
x_eg=0.30 #mass fraction
total=m_solvent+m_solution
percent_eg=x_eg*m_solution/total*100
percent_w=(1-x_eg)*m_solution/total*100
percent_solvent=m_solvent/total*100
print('% ethylene glycol=',percent_eg)
print('% water=',percent_w)
print('% furfural=',percent_solvent)
```

Extract is 89% furfural, 7% ethylene glycol, 4% water.

Raffinate is 9% furfural, 24% ethylene glycol, 67% water.

```python
x_f_E=0.89
x_eg_E=0.07
x_w_E=0.04
x_f_R=0.09
x_eg_R=0.24
x_w_R=0.67

EM=3.88 #in
ER=7.86 #in
MR=ER-EM #in

m_E=MR/ER*(m_solvent+m_solution)
m_R=(m_solvent+m_solution)-m_E
m_eg_E=x_eg_E*m_E
m_eg_R=x_eg_R*m_R
print('amount of ethylene glycol in the extract=',m_eg_E,'kg')
print('amount of ethylene glycol in the raffinate=',m_eg_R,'kg')
print('amount of solvent free extract=',m_E*(1-x_f_E),'kg')
print('amount of solvent free raffinate=',m_solution-(m_E*(1-x_f_E)),'kg')
percent_extracted=m_eg_E/(x_eg*m_solution)*100
print('% ethylene glycol extracted=',percent_extracted)
```

The amount of solvent-free extract is $\boxed{5.01 \space kg}$.

The amount of solvent-free raffinate is $\boxed{39.99 \space kg}$.

The percentage of ethylene glycol extracted is $\boxed{23.6 \% }$.
