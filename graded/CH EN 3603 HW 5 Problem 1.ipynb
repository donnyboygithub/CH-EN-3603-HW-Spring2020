## Part 1

The constants $A$, $M_A$, $M_B$, $p$, $\sigma_{AB}$, and $\Omega$ in the Chapam-Enskog expression can be grouped into one constant $K$, which leads to

$D_{AB}=K T^{3/2}$.

Using the values for $D_{AB}$ and $T$ from homework 4, we can solve for K using

$K=\frac{D_{AB}}{T^{3/2}}$.

```python
T_C=25 #degrees C
T_K=T_C+273.15 #K
D_AB=0.0905 #cm^2/s

K=D_AB/(T_K)**(3/2)

print("K=",K,"cm^2/(s K^(3/2))")
```

```python
from numpy import linspace
import matplotlib.pyplot as plt

TC=linspace(10,80,1000)
TK=273.15+TC
DAB=K*TK**(3/2)

plt.plot(TC,DAB)
plt.title('D_AB vs Temperature')
plt.xlabel('Temperature (degrees C)')
plt.ylabel('D_AB (cm^2/s)')
plt.grid()
plt.show()
```
