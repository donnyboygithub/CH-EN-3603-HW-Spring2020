The separation of isopentane from n-pentane by distillation is difficult (approximately 100 trays are required),

# Solution

## part 1

The Antoine vapor pressure equation is

$\log_{10}(p)=A-\frac{B}{T+C}$

with pressure given in bar and temperature given in Kelvin.

```python
c=14.5038 #conversion from bar to psi

#for iC5
Ai=3.97183
Bi=1021.864
Ci=-43.231
#source: https://webbook.nist.gov/cgi/cbook.cgi?ID=C78784&Mask=4

#for nC5
An=3.9892
Bn=1070.617
Cn=-40.454
#source: https://webbook.nist.gov/cgi/cbook.cgi?ID=C109660&Mask=4&Type=ANTOINE&Plot=on
```

```python
import matplotlib.pyplot as plt
import numpy as np

TF=np.array([125,150,175,200,225,250]) #F
TK=(TF-32)*(5/9)+273.15 #K
Pi=c*10**(Ai-Bi/(TK+Ci)) #psia
Pn=c*10**(An-Bn/(TK+Cn)) #psia

plt.plot(TF,Pi,'b-',label='isopentane')
plt.plot(TF,Pn,'r-',label='n-pentane')
plt.grid()
plt.legend(loc='best')
plt.xlabel('Temperature (degrees F)')
plt.ylabel('Saturation Pressure (psia)')
plt.show()
```

## part 2

To derive an equation for making predictions using Roult's law, the following equation are used, and the following steps are taken:

$K_i=\frac{y_i}{x_i}=\frac{P_i^s}{P}$  (this is a form of Roult's law)

$\alpha_{in}=\frac{K_i}{K_n}=\frac{P_i^s/P}{P_n^s/P}=\frac{P_i^s}{P_n^s}$

```python
alpha=np.array([1.26,1.23,1.21,1.18,1.16,1.14])
predictions=Pi/Pn

plt.plot(TF,alpha,'bo',label='alpha')
plt.plot(TF,predictions,'r-',label='predictions from Roult\'s Law')
plt.legend(loc='best')
plt.grid()
plt.xlabel('Temperature (degrees F)')
plt.ylabel('alpha (unitless)')
plt.show()
```

We can conclude that for this binary system, in this temperature range, predictions from Roult's law are not accurate enough, and will become increasingly less accurate with time, as the trend from the graph indicates.
