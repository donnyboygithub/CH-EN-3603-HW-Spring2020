A feed of 13,500 kg/h is 8 wt% acetic acid (B) in water (A). Removal of acetic acid is to be by liquid–liquid extraction at 25°C, as shown schematically in Figure 1. The raffinate is to contain 1 wt% acetic acid. The following four solvents, with accompanying distribution (partition) coefficients in mass-fraction units, are candidates. $K_{D_B}=\frac{\omega_B^{(E)}}{\omega_B^{(R)}}, \space\space\space\space\space\space\space (1)$
