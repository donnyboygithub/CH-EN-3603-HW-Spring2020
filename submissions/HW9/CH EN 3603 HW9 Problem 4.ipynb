# Problem 4

An $SO_2air$ mixture is scrubbed with water in a packed tower at 20C and 1 atm. Solute-free water enters the top at 1,000 lb/h and is well distributed over the packing. The liquid leaving contains 0.6 lb $SO_2$ per 100 lb of solute-free water. The partial pressure of $SO_2$ in the gas leaving is 23 torr. The mole ratio of water to air is 25. The necessary equilibrium data are tabulated below.



1. What percent (on a mole basis) of the $SO_2$ in the entering gases is absorbed in the tower?

2. During operation, it was found that rate coefficients $k_L$ and $k_p$ remained substantially constant throughout the tower at $k_L=1.3 \frac{ft}{hr}$ and $k_p=0.195 \frac{lbmol}{hr \cdot ft^2 \cdot atm}$. At a point in the tower where the liquid concentration is 0.001 lbmol $SO_2$ per lbmol of water, what is the liquid concentration at the gasliquid interface (in lbmol/$ft^3$) and $SO_2$ vapor partial pressure at the interface (in atm)? The solution density is 1 g/cm$^3$.

## part 1 solution

The fraction of $SO_2$ that is absorbed in the tower can be found by finding the molar flow rates of $SO_2$ in the outlet gas and in the outlet liquid. The sum of these values will be equal the molar flow rate of $SO_2$ entering in the gas. Then, the ratio of the molar flow rate of $SO_2$ leaving in the liquid to the molar flow rate of $SO_2$ entering in the gas will equal the fraction of $SO_2$ that was absorbed in the tower.

The molar flow rate of $SO_2$ in the outlet gas is found by first calculating to molar flow rate of air in the system. This can be done using the mole ratio of water to air.

### $\dot n_{air}=\frac{\dot m_{water}}{MW_{water}(ratio \space of \space water \space to \space air)}$

```python
m_water=1000 #lb-water/hr
MW_water=18.01528 #lb-water/lb-mol-water
ratio=25 #lb-mol-water/lb-mol-air

n_air=m_water/(MW_water*ratio)
print('n_air=',n_air,'lb-mol-air/hr')
```

Since the partial pressure of $SO_2$ in the exiting gas stream is known, and the total pressure is known, the partial pressure of air in the exiting gas stream can be calculated.

### $p_{air}=P-p_{SO_2}$

```python
p_SO2=23 #torr
P=760

p_air=P-p_SO2
print('p_air=',p_air,'torr')
```

The mole fraction of air in the exiting gas stream can then be calculated with the equation

### $y_{air}=\frac{p_{air}}{P}$

```python
y_air=p_air/P
print('y_air=',y_air)
```

From this, the total molar flow rate of the exiting gas stream can be calculated with the equation

### $\dot n_T=\frac{\dot n_{air}}{y_{air}}$

```python
n_T=n_air/y_air
print('n_T=',n_T,'lb-mol/hr')
```

The molar flow rate of $SO_2$ in the exit gas stream can then be calculated with the equation

### $\dot n_{SO_2}=y_{SO_2}\cdot \dot n_T=(1-y_{air})\cdot \dot n_T$

```python
n_SO2_gas=(1-y_air)*n_T
print('n_SO2_gas=',n_SO2_gas,'lb-mol-SO2/hr')
```

The molar flow rate of $SO_2$ in the outlet liquid can be found by first finding the mass flow rate of $SO_2$ in the exit liquid stream. This is done with the equation

### $\dot m_{SO_2}=solubility \cdot \dot m_{water}$

where solubility equals the mass of $SO_2$ per mass of water.

```python
solubility=0.006 #lb-SO2/lb-water

m_SO2=solubility*m_water
print('m_SO2=',m_SO2,'lb-SO2/hr')
```

This is then converted into the molar flow rate of $SO_2$ in the exit liquid stream with the equation

### $\dot n_{SO_2}=\frac{\dot m_{SO_2}}{MW_{SO_2}}$

```python
MW_SO2=64.066 #lb-SO2/lb-mol-SO2

n_SO2_liquid=m_SO2/MW_SO2 #lb-mol-SO2/hr
print('n_SO2_liquid=',n_SO2_liquid,'lb-mol-SO2/hr')
```

The inlet molar flow rate of $SO_2$ in the gas stream can now be calculated by summing the molar flow rate of $SO_2$ in the exit gas stream and in the exit liquid stream.

```python
n_SO2_in=n_SO2_gas+n_SO2_liquid
print('n_SO2_in=',n_SO2_in,'lb-mol-SO2/hr')
```

The percentage of $SO_2$ in the entering gases that was absorbed in the tower is then

### $percentage=\frac{\dot n_{SO_2} \space in \space exiting \space liquid \space stream}{\dot n_{SO_2} \space in \space entering \space gas \space stream} \times 100\% $

```python
percentage=n_SO2_liquid/n_SO2_in*100
print('percentage=',percentage,'%')
```

The percent (on a mole basis) of the $SO_2$ in the entering gases that was absorbed in the tower is $\boxed{57.5\%}$.

## part 2 solution

The molar flux in the gas phase is

$N_{SO_2}=k_p(p_{SO_2,B}-p_{SO_2,I})$

where $p_{SO_2,B}$ is the partial pressure of $SO_2$ in the bulk gas, and $p_{SO_2,I}$ is the partial pressure of $SO_2$ at the interface.

The molar flux in the liquid phase is

$N_{SO_2}=k_L(c_{SO_2,I}-c_{SO_2,B})$

where $c_{SO_2,B}$ is the concentration of $SO_2$ in the bulk liquid, and $c_{SO_2,I}$ is the concentration of $SO_2$ at the interface.

At equilibrium, the molar fluxes are equal $\Rightarrow$

$k_p(p_{SO_2,B}-p_{SO_2,I})=k_L(c_{SO_2,I}-c_{SO_2,B}) \Rightarrow$

### $\frac{k_p}{k_L}=\frac{(p_{SO_2,B}-p_{SO_2,I})}{(c_{SO_2,I}-c_{SO_2,B})}$

When the liquid concentration is 0.001 lbmol $SO_2$ per lbmol of water, the molar flow rates of $SO_2$ and water are related by

### $\frac{0.001 \space lbmol \space SO_2}{lbmol \space water}=\frac{\dot n_{SO_2}}{\dot m_{water}/MW_{water}} \Rightarrow$

### $\dot n_{SO_2}=\frac{0.001 \space lbmol \space SO_2}{lbmol \space water} \cdot \frac{\dot m_{water}}{MW_{water}}$

```python
n_SO2_L=0.001*m_water/MW_water
print('n_SO2_L=',n_SO2_L,'lbmol-SO2/hr')
```

Because $SO_2$ in the gas + $SO_2$ in the entering liquid = $SO_2$ in the liquid + $SO_2$ in the exiting gas, this leads to

```python
n_SO2_G=n_SO2_L+n_SO2_gas
print('n_SO2_G=',n_SO2_G,'lbmol-SO2/hr')
```

Because the molar flow rate of air is constant throughout the tower, the partial pressure of $SO_2$ in the bulk gas is

### $p_{SO_2,B}=\frac{\dot n_{SO_2,G}}{\dot n_{SO_2,G}+\dot n_{air}}$



```python
p_SO2_B=n_SO2_G/(n_SO2_G+n_air)
print('p_SO2_B=',p_SO2_B,'atm')
```

The concentration of $SO_2$ in the bulk liquid can be found with the equation

### $c_{SO_2,B}=\frac{\left ( \frac{0.001 lbmol-SO_2}{lbmol-water}\right )}{MW_{water}} \times \rho_{water}$

```python
_water=62.4 #lb-water/ft^3

c_SO2_B=0.001/MW_water*_water
print('c_SO2_B=',c_SO2_B,'lbmol-SO2/fr^3')
```

The solubility data for $SO_2$ in water is converted from a mass basis to a mole basis by using the molecular weights.

```python
import numpy as np

mass_solubility_L_100=np.array([0.02,0.05,0.10,0.15,0.20,0.30,0.50,0.70,1.0]) #lb-SO2/lb-water
mole_solubility_L=mass_solubility_L_100*MW_water/MW_SO2
print(mole_solubility_L,'lbmol-SO2/100 lbmol-water')
```

The concentration of $SO_2$ at the various solubilities is then calculated using

### $c_{SO_2,I}=\frac{(molar \space solubility \space of \space SO_2 \space in \space water)}{MW_{water}}\times \rho_{water}$

```python
c_SO2_I=mole_solubility_L/MW_water*_water/100
print('c_SO2_I=',c_SO2_I,'lbmol-SO2/ft^3')
```

The partial pressure equilibrium data is then converted into units of atmospheres.

```python
partial_pressure_V_atm=np.array([0.5,1.2,3.2,5.8,8.5,14.1,26.0,39.0,59])/760 #atm
print('partial_pressure_V_atm=',partial_pressure_V_atm,'atm')
```

The concentration data is then plotted against the partial pressure data.

```python
import matplotlib.pyplot as plt

#plt.figure(figsize=(20,10))
plt.plot(c_SO2_I,partial_pressure_V_atm)
plt.grid()
plt.ylabel('p_SO2 (atm)')
plt.xlabel('c_SO2 (lbmol-SO2/ft^3)')
plt.show()
```

A correlation between the concentration data and the partial pressure data is created using the 'polyfit' function.

```python
def Fc(c_i):
    cv=np.polyfit(c_SO2_I,partial_pressure_V_atm,3)
    c = cv[0]*c_i**3 + cv[1]*c_i**2 + cv[2]*c_i + cv[3]
    return c
```

With the equation derived above,

### $\frac{k_p}{k_L}=\frac{(p_{SO_2,B}-p_{SO_2,I})}{(c_{SO_2,I}-c_{SO_2,B})}$

$p_{SO_2,I}$ is replaced with the fitted equation,

$p_{SO_2,B}$ is replaced with 0.0532 atm,

$c_{SO2,B}$ is replaced with 0.00346 lbmol-SO2/fr^3 $\Rightarrow$

### $0=\frac{k_p}{k_L}-\frac{\left (0.0532 atm-p_{SO_2,I}(c_{SO_2,I}) \right )}{\left (c_{SO_2,I}-0.00346 \right )}$

This residual is then solved using the 'fsolve' function to find a value for $c_{SO_2,I}$

```python
#p_SO2_B=0.0532 #atm
#c_SO2_B=0.00346 #lbmol-SO2/fr^3

k_p=0.195 #lbmol/(hr*ft^2*atm)
k_L=1.3 #ft/hr

from scipy.optimize import fsolve

def func(c_i):
    r=k_p/k_L-(p_SO2_B-Fc(c_i))/(c_i-c_SO2_B)
    return r

sol=fsolve(func,0.008)
print(sol,'lbmol/ft^3')
```

Plugging $c_{SO_2,I}=0.00695 \frac{lbmol}{ft^3}$ back into the equation 

### $\frac{k_p}{k_L}=\frac{(p_{SO_2,B}-p_{SO_2,I})}{(c_{SO_2,I}-c_{SO_2,B})} \Rightarrow$

```python
p_SO2_I=p_SO2_B-k_p/k_L*(sol-c_SO2_B)
print(p_SO2_I,'atm')
```

The liquid concentration at the gas-liquid interface is $\boxed{0.06953 \space \frac{lbmol}{ft^3}}$.

The $SO_2$ vapor partial pressure at the interface is $\boxed{0.0527 \space atm}$.
